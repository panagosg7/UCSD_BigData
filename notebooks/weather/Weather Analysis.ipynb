### Computing mean and std for each station/year ###

```python
# a simple scale function to normalize the data-frame row-by-row
from numpy import mean, std
def scale_temps(Din):
    matrix=Din.iloc[:,3:]
    Dout=Din.loc[:,['station','year']+range(1,366)]
    Mean=mean(matrix, axis=1).values
    Dout['Mean']=Mean
    Std= std(matrix, axis=1).values
    Dout['Std']=Std
    # Decided not to normalize each year to have mean zero and std 1
    # tmp = pd.DataFrame((matrix.values - Mean[:,np.newaxis])/Std[:,newaxis],columns=range(1,366))
    # print tmp.head()
    Dout.loc[:,1:365]=matrix.values
    return Dout
Dout=scale_temps(Data)
#reorder the columns
Dout=Dout[['station','year','Mean','Std']+range(1,366)]
Dout.head()
```

### NaN-tolerant averaging  ###
We compute the empirical covariance matrix in a way that tolerates NaN values.

<span style="color:red"> In the code below I remove all rows that have a nan in them. If you remve the command **M.dropna(...** then all rows are used. Can you get better results without removing the rows? </span>

```python
# demonstrating the use of the cell magic %%time, which measures the run-time of the cell.
M=Dout.loc[:,1:365].transpose()
M=M.dropna(axis=1)
(columns,rows)=shape(M)
Mean=mean(M, axis=1).values

print (columns,rows), shape(Mean)
C=np.zeros([columns,columns])   # Sum
N=np.zeros([columns,columns])   # Counter of non-nan entries
```

```python
%%time
for i in range(rows):
    if i % 1000==0: 
        print i
    row=M.iloc[:,i]-Mean;
    outer=np.outer(row,row)
    valid=isnan(outer)==False
    C[valid]=C[valid]+outer[valid]  # update C with the valid location in outer
    N[valid]=N[valid]+1
valid_outer=np.multiply(1-isnan(N),N>0)
cov=np.divide(C,N)
```

```python
shape(cov)
```

```python
U,D,V=np.linalg.svd(cov)
```

```python
shape(U),shape(D),shape(V)
```

```python
## AUXILIARY FUNCTIONS FOR LATER ON:


############## COLOR CREATION ##############
# Taken from here: http://bensouthgate.com/p/12_3_13.php


def hex_to_RGB(hex):
    ''' "#FFFFFF" -> [255,255,255] '''
    # Pass 16 to the integer function for change of base
    return [int(hex[i:i+2], 16) for i in range(1,6,2)] 


def RGB_to_hex(RGB):
    ''' [255,255,255] -> "#FFFFFF" '''
    # Components need to be integers for hex to make sense
    RGB = [int(x) for x in RGB]
    return "#"+"".join(["0{0:x}".format(v) if v < 16 else 
                        "{0:x}".format(v) for v in RGB])
    


def color_dict(gradient):
    ''' Takes in a list of RGB sub-lists and returns dictionary of 
        colors in RGB and hex form for use in a graphing function 
        defined later on '''
    return {"hex":[RGB_to_hex(RGB) for RGB in gradient],
            "r":[RGB[0] for RGB in gradient],
            "g":[RGB[1] for RGB in gradient],
            "b":[RGB[2] for RGB in gradient]}


def linear_gradient(start_hex, finish_hex="#FFFFFF", n=10):
    ''' returns a gradient list of (n) colors between two hex colors.
        start_hex and finish_hex should be the full six-digit color string, 
        inlcuding the number sign ("#FFFFFF") '''
    # Starting and ending colors in RGB form
    s = hex_to_RGB(start_hex)
    f = hex_to_RGB(finish_hex)
    # Initilize a list of the output colors with the starting color
    RGB_list = [s]
    # Calcuate a color at each evenly spaced value of t from 1 to n
    for t in range(1, n):
        # Interpolate RGB vector for color at the current value of t
        curr_vector = [ int(s[j] + (float(t)/(n-1))*(f[j]-s[j])) 
                        for j in range(3)]
        # Add it to our list of output colors
        RGB_list.append(curr_vector)

    return color_dict(RGB_list)

##########################################
```

```python
# http://matplotlib.org/basemap/users/merc.html

from mpl_toolkits.basemap import Basemap, cm
import numpy as np
import matplotlib.pyplot as plt

### INPUTS ################

temp_step=4       # temperature step - how to bunch up temperatures to colors

year=2000

###########################


plt.figure(figsize=(15,10),dpi=300)
m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
m.drawcoastlines()
m.fillcontinents(color='#EFEFEF',lake_color='aqua')

## draw parallels and meridians.
#parallels = np.arange(-80,81,10.)
#labels = [left,right,top,bottom]
#m.drawparallels(parallels,labels=[False,True,True,False])
#meridians = np.arange(10.,351.,20.)
#m.drawmeridians(meridians,labels=[True,False,False,True])

#m.drawparallels(np.arange(-90.,91.,30.))
#m.drawmeridians(np.arange(-180.,181.,60.))
m.drawmapboundary(fill_color='aqua')

#############################
vals=Djoined.ix[:,['longitude', 'latitude', 'Mean','year']].values


# Group up points by temperature
def temp_groups(min_temp, max_temp, step):
    # Cut temperatures that are probably wrong:
    if max_temp > 50:
        max_temp=50    
    if min_temp < -40:
        min_temp=-40    
    
    # Compute all points in the range of positive temperatures
    pos_range=[ t for t in xrange(0,int(max_temp),step)]
    ps=zip(pos_range[:len(pos_range)-1], pos_range[1:len(pos_range)])    
    warm_colors=linear_gradient("#FFD3C6","#CC0000",len(ps))['hex']
    #print warm_colors
    
    warm_points = [ (l,h,c, [ (x,y) for (x,y,t,yr) in vals if year==yr and l<=t and t<h ]) 
                    for ((l,h),c) in zip(ps,warm_colors)]
    
    # Compute all points in the range of negative temperatures
    neg_range=[ t for t in xrange(0,int(min_temp), (-1)*step)]
    neg_range.reverse()
    ns=zip(neg_range[:len(neg_range)-1], neg_range[1:len(neg_range)])
    cold_colors=linear_gradient("#000099","#B7D9FF",len(ns))['hex']    
    #print cold_colors
    cold_points = [ (l,h,c, [ (x,y) for (x,y,t,yr) in vals if year==yr and l<=t and t<h ]) 
                    for ((l,h),c) in zip(ns,cold_colors)]
    
    return cold_points + warm_points
    
    
temps=[t for (x,y,t,yr) in vals ]
min_temp=min(temps)
max_temp=max(temps)
        
# groups: all stations grouped based on temperature
tgroups=[ g for g in temp_groups(min_temp,max_temp,temp_step) ]

# Add the points to the graph
for (l,h,c,g) in tgroups:
    xs=[x for (x,y) in g]
    ys=[y for (x,y) in g]
    print len(xs), ' points with temp ', l,h
    x, y = m(xs,ys)
    m.plot(x,y,color=c, linestyle='None', marker='.',label=str(l) + ' - ' + str(h))

plt.title('Average temperature (in C) for the year ' + str(year))
plt.legend(loc=3)
plt.show()
```

### Reconstruction ###

```python
def plot_reconstructions(selection,rows=2,columns=7):
    Recon=Eig*Prod.transpose()+Mean[:,np.newaxis]
    plt.figure(figsize=(columns*3,rows*3),dpi=300)
    j=0;
    for i in selection:
        subplot(rows,columns,j); 
        j += 1; 
        if j>=rows*columns: break
        plot(Recon[:,i])
        plot(Djoined.ix[i,1:365]);
        title(Djoined.ix[i,'station']+' / '+str(Djoined.ix[i,'year']))
        xlim([0,365])
```
