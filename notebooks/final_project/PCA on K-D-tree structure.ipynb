## Merging: main iteration

```python
# active_data :: { id: (n, mean, cov) }
active_data = read_mr_results(cmd_id, output_file)
done = False
itr = 1

print '--------------------------------'
print 'Initial global map'
print '--------------------------------'

hi.plot_map()

while not done:   
    
    print '================================'
    print
    print 'Iteration', itr
    print
    print '================================'    

    hi.remove_groups(list(set(hi.get_groups()) - set(active_data.keys())))
    
    (failed, old_groups) = do_PCA(active_data, False)
    
    for f in failed:
        print 'Removing from active_data failed group', f    
        del active_data[f]    
    
    hi.remove_groups(failed)
    
    # make pairs and calculate new group statistics
    pairs = make_pairs(hi,active_data.keys())
    (potential_pairs, new_groups, new_data) = new_group_stats(cmd_id, old_groups, pairs)

    # apply the criterion
    accepted_pairs = mld_criterion(old_groups, new_groups, potential_pairs)
    print 'Merging', len(accepted_pairs), 'groups'

    for (k,i,j) in accepted_pairs:
        hi.merge_groups(k,i,j)
        # remove old bindings
        del active_data[i]
        del active_data[j]
        #add new group to the data store
        active_data[k] = new_data[k]

    print '--------------------------------'
    print 'Initial global map'
    print '--------------------------------'
    hi.plot_map()

    print '--------------------------------'
    print 'Explained variance after merge'
    print '--------------------------------'
    plot_expl_var(active_data)
    
    if len(active_data.keys()) < 2:
        done = True

    if len(accepted_pairs) == 0:
        done = True    
    itr = itr + 1
```
