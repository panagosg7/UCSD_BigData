```python
%%writefile hierarchical_indexing_1.py
import pandas as pd
import numpy as np
import sklearn as sk
print 'pandas version: ',pd.__version__
print 'numpy version:',np.__version__
print 'sklearn version:',sk.__version__


# Polygon imports - careful to use diffrerent names than matplotlib Polygon
import Polygon as pg
import Polygon.IO as pgio
from Polygon.Shapes import Rectangle
from Polygon.Utils import convexHull, tile, tileEqual, tileBSP, reducePoints, cloneGrid, prunePoints
import random, math

class HierarchicalIndexing:
    
    # Pandas.DataFrame indexed by station ID, containing at least the cilumns:
    # * weight
    # * longitude
    # * latitude
    _orig_df          = None
    
    # int: levels of hierarchy tree
    _levels           = 0
    
    # a new Pandas.DataFrame containing station information including the 'group_id' 
    _grouped_stations = None
    
    # Dictionary with entries of the form:
    # { 
    #   '<gid>' : (weight, Polygon)
    # }
    _groups           = None           
    
    # boolean: True if the analysis has run, False otherwise
    _computed         = False
    
    # np.ndarray with dimensions: (2**levels, 2**levels)
    _neighbors        = None
    
    # a counter for fresh gid's
    _gid_counter      = 0
    
    
    
    def __init__(self, df, levels):
        """
        Initializer

        Parameters
        ----------
        df    : Pandas.DataFrame
                Indexed by station ID
                Required fields:
                 * weight
                 * longitude
                 * latitude

        level : int
                  Level of iteration - reasonable values are > 0 
        """        
        
        # set level
        self._levels = levels
        
        # keep a copy of the original input dataframe
        self._orig_df = df


    def get_groups(self):
        if not self._computed:
            self.run()
        return self._groups

    
    def get_stations(self):
        if not self._computed:
            self.run()
        return self._grouped_stations

    
    def next_c(self,c):
        if   c == 'longitude':
            return 'latitude'
        elif c == 'latitude':
            return 'longitude'
        else:
            raise Exception('long/lat')

            
    def level_str(self,i):
        return 'g' + str(i)
    
    
    def fresh_gid(self):
        num = self._gid_counter
        self._gid_counter = num + 1
        return num
        
        
        
        
    ## Polygon
    
    def polygon_to_lists(self,poly):
        """
        Given a polygon with nodes: (x1,y1), ..., (xn,yn), 
        `polygon_to_lists` returns:
        ([x1,..,xn], [y1,...,yn])
        
        """
        return tuple(map(list, zip(*poly[0])))
        
    
    
    ## Compute Neighbors 
    
    def __neighboring__(self,g1,g2):
        """
        Returns True if group `i` is neighboring with group `j`, 
        otherwise returns False.
        
        TODO: wrap around the globe
        
        """        
        _,p1 = self._groups[g1]
        _,p2 = self._groups[g2]
        print g1,g2
        print p1
        print p2
        union = prunePoints(p1+p2)
        print union
        print
        
        result = (len(union) == 1)
        return result
    
    def __compute_neighbors__(self):
        if not self._computed:
            self.run()
                
        group_no = len(self._groups)
        
        self._neighbors = np.zeros((group_no,group_no), dtype=np.bool)
        
        for i in range(0,group_no):
            for j in range(i,group_no):
                if self.__neighboring__(i,j):
                    ## Add edge
                    self._neighbors[i][j] = True
                    self._neighbors[j][i] = True
        
    
    def get_neighbors_of(self,n):
        """       
        For a give group_id `n`, returns a list of the neighboring groups.
        """
        narr = self.get_neighbors()        
        group_no = len(self._groups)
        for j in range(0,group_no):         
            if narr[n][j]:
                yield j
    
    
    def are_neighbors(self,i,j):
        """       
        Given two groups `i` and `j`, returns True if these groups are neighboring,
        False otherwise.
        """
        narr = self.get_neighbors()
        return narr[i][j]            
    
                
    def get_neighbors(self):
        if self._neighbors == None:
            self.__compute_neighbors__()
        return self._neighbors
    


    # Top-level runner 
    def run(self):        
        if not self._computed:
            # make a deep copy every time `run` is called
            df = self._orig_df.copy(deep=True)
            
            # run the analysis
            (self._grouped_stations, group_list) = self.__run__(df, self._levels)            
            
            columns = ['group_id', 'weight', 'polygon'] # 'west', 'east', 'south', 'north']
            
            # populate the _groups dictionary
            self._groups = {}    
            for g in group_list:
                gid = g['group_id']
                w = g['weight']
                poly = g['poly']
                self._groups[gid] = (w,poly)
        
            self._gid_counter = len(self._groups)
            self._computed = True            

            
    def path_to_key(self, a): 
        return sum([ x * (2**i) for (x,i) in zip(a, range(0, len(a)))])
            
        
    def __run__(self, df, level, coord='latitude', west=-180, east=180, south=-90, north=90, path=[]):
        """
        __run__

        Parameters
        ----------
        self    :
        
        df      : Pandas.DataFrame
                  
        level   : int
                  Level of iteration

        coord   : str ('longitide' or 'latitude')
                  The first time this is run it really should be 'latitude' so that we 
                  establish the main separation of northern and southern hemisphere.

        west    : int
                  the western boundary of the region

        east    : int
                  the eastern boundary of the region

        south   : int
                  the southern boundary of the region

        north   : int
                  the northern boundary of the region
                  
        path    : list of boolean
                  the branches taken so far in the hierarchy index tree

               if coord = 'longitude':

                       north
              ______________________________
             |FALSE                 |   TRUE|   (90)
             |               new -> |       |
        west |             boundary |       | east
             |                      |       | 
             |______________________|_______|   (-90) 
           (-180)       south             (180)   


                if coord = 'latitude':      

                       north
              _______________________
             |TRUE           new     |  (90)
             |             boundary  |
        west |                |      | east
             |________________v______|
             |FALSE                  |
             |                       |
             |_______________________|  (-90) 
           (-180)       south      (180)   


        Returns
        -------
        A new dataframe containing the columns of the original dataframe,
        extended with columns g1, g2, ... that altogether denote the group
        that each station belongs to.
        """
        
        if level <= 0:
            # we have reached a leaf:
            # * compute the group id
            # * update the stores
            group_id = self.path_to_key(path)
            
            rect = Rectangle(1)
            rect.warpToBox(west,east,south,north)
            
            gr = { 
                   'group_id' : group_id,
                   'weight'   : df['weight'].sum(),
                   'poly'     : rect                   
                 }
            
            df['group_id'] = group_id
            return (df,[gr])

        # Sort by coord
        df = df.sort(coord)

        # Compute a new column with the cummulative count of 
        df['cum_weight'] = df['weight'].cumsum()

        # compute the threshold of weight (the point that designates the meridian)
        threshold = df['weight'].sum() / 2

        # group index column
        gi = self.level_str(level)

        # assign a group number based on weight
        df[gi] = df['cum_weight'] > threshold

        # drop the 'cum_weight' column
        del df['cum_weight']

        # break down into groups based on the grouping column
        gdf = df.groupby([gi])
        
        # ugly way to get the boundary of the partition on the coordinate of interest
        new_boundary = np.average([[group[coord].max() for name, group in gdf if name==False][0], \
                                   [group[coord].min() for name, group in gdf if name==True ][0]])

        # recursively partition the groups
        children_df = []
        children_gr = []


        for name, child in gdf:
            
            new_east = east
            new_west = west
            new_north = north
            new_south = south
            
            if name==False: 
                # cum_weight is smaller than threshold, i.e.
                #   - station is west of the median if it's longitude, so east gets updated
                #   - station is south of the median if it's latitude, so north gets updated
                if coord=='longitude':
                    new_east = new_boundary
                    #print coord, name, west,'<=', new_east,'<=', east
                else:
                    new_north = new_boundary
                    #print coord, name, south,'<=', new_north,'<=', north

            else:            
                # cum_weight is greater than threshold, i.e.
                #   - station is east of the median if it's longitude, so west gets updated
                #   - station is north of the median if it's latitude, so south gets updated
                if coord=='longitude':
                    new_west = new_boundary
                    #print coord, name, west,'<=', new_west,'<=', east

                else:
                    new_south = new_boundary
                    #print coord, name, south,'<=', new_south,'<=', north


            #print 'parent', path, 'coords:', east, west, south, north             
            #print '  ', 'branch', name, new_east, new_west, new_south, new_north 
            #print
            
            #def between(x, left, right): 
            #    if (left <= x) and (x <= right):
            #        return True
            #    else:
            #        #print left, x, right
            #        return False            
            
            #if not between(new_east, west, east): 
            #    print 'error 1'
            #if not between(new_west, west, east): 
            #    print 'error 2'
            #if not between(new_north, south, north): 
            #    print 'error 3'
            #if not between(new_south, south, north): 
            #    print 'error 4'
            
            
            # do the recursive call using updates boundaries
            # ignore_index = True
            (child_df, child_gr) = self.__run__(child,
                                                level-1,
                                                self.next_c(coord),
                                                new_west,new_east,new_south,new_north,
                                                [int(name)]+path)
            children_df.append(child_df)
            children_gr=children_gr + child_gr

        # concatenate children and return
        return (pd.concat(children_df), children_gr)
```
