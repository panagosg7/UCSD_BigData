##Plot group map

The following functions creates a polygon (in our case a rectangle
that corresponds to the input longitudes (`west`, `east`) and latitudes 
(`south`, `north`). Each group is labeled by the group id `gid`.

```python
def draw_screen_poly(plt, gid, west, east, south, north, m, color, text=True):
    """
    Parameters
    ----------
    plt     : the map plot
    gid     : group's id
    west    : west border  
    east    : east border 
    south   : south border
    north   : north border
    m       : basemap
    color   : color of the rectangle
    text    : print group_id tags on each group on the map (default: True)
    
    Effect
    ------
    Renders a `color` colored rectangle on the map
    """
    def fix(x):
        if x >  80: return 80.0
        if x < -80: return -80.0
        return x
    
    lons = [west, east, east, west]
    lats = map(fix, [south, south, north, north])
    x, y = m( lons, lats )
    xy = zip(x,y)
    
    poly = Polygon( xy, facecolor=color, alpha=0.4, label = str(gid) )

    
    if text:
        # center    
        clon = (west + east) / 2
        clat = fix((north + south) / 2)
        cx,cy = m([clon],[clat])
        plt.text(cx[0], cy[0], str(gid), 
                 horizontalalignment='center', 
                 verticalalignment='center',
                 bbox=dict(facecolor='white', alpha=1),
                 size =15, fontweight='bold')  
    
    
    plt.gca().add_patch(poly)
```

```python
# Taken from here: http://bensouthgate.com/p/12_3_13.php

def hex_to_RGB(hex):
    ''' "#FFFFFF" -> [255,255,255] '''
    # Pass 16 to the integer function for change of base
    return [int(hex[i:i+2], 16) for i in range(1,6,2)] 


def RGB_to_hex(RGB):
    ''' [255,255,255] -> "#FFFFFF" '''
    # Components need to be integers for hex to make sense
    RGB = [int(x) for x in RGB]
    return "#"+"".join(["0{0:x}".format(v) if v < 16 else 
                        "{0:x}".format(v) for v in RGB])
    


def color_dict(gradient):
    ''' Takes in a list of RGB sub-lists and returns dictionary of 
        colors in RGB and hex form for use in a graphing function 
        defined later on '''
    return {"hex":[RGB_to_hex(RGB) for RGB in gradient],
            "r":[RGB[0] for RGB in gradient],
            "g":[RGB[1] for RGB in gradient],
            "b":[RGB[2] for RGB in gradient]}


def linear_gradient(start_hex, finish_hex="#FFFFFF", n=10):
    ''' returns a gradient list of (n) colors between two hex colors.
        start_hex and finish_hex should be the full six-digit color string, 
        inlcuding the number sign ("#FFFFFF") '''
    # Starting and ending colors in RGB form
    s = hex_to_RGB(start_hex)
    f = hex_to_RGB(finish_hex)
    # Initilize a list of the output colors with the starting color
    RGB_list = [s]
    # Calcuate a color at each evenly spaced value of t from 1 to n
    for t in range(1, n):
        # Interpolate RGB vector for color at the current value of t
        curr_vector = [ int(s[j] + (float(t)/(n-1))*(f[j]-s[j])) 
                        for j in range(3)]
        # Add it to our list of output colors
        RGB_list.append(curr_vector)

    return color_dict(RGB_list)
```

##PCA - inline

```python
%%writefile pca.py
#!/usr/bin/python

"""
<description here>
"""

import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr
import gzip
import pickle
import numpy as np
import pandas as pd

class MRPCA(MRJob):

    def configure_options(self):
        super(MRPCA,self).configure_options()
        self.add_file_option('--stations')
    
    def mapper_init(self):
        f = gzip.open( self.options.stations, "rb" )
        pickleFile = pickle.Unpickler( f )
        self.stations = pickleFile.load()[['latitude','longitude','group_id']]
        f.close()
    
    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper-all',1)
            elements = line.split(',')
            
            # only include TMIN measurements
            if elements[1] == 'TMAX':

                st_id = elements[0]

                # the group_id of the station in the current 
                group_id = int(self.stations.loc[st_id]['group_id'])

                # pack just the content back into an comma separated string
                yield (group_id, ','.join(elements[3:]))
        
        except KeyError as e:
            stderr.write('Key was not found.\n')
            
        except Exception as e:
            stderr.write('Error in line:\n'+line)
            stderr.write(str(e))
            self.increment_counter('MrJob Counters','mapper-error',1)
            
            
#     def combiner(self, key, values):
#         self.increment_counter('MrJob Counters','combiner',1)                
#         yield (key, values)


    # PCA is run on the reducer
    
    def reducer(self, group_id, values):
        self.increment_counter('MrJob Counters','reducer',1)        

        def to_int(s):
            if s=='': 
                return np.NaN 
            else: 
                return int(s)
        
        vrows = [ v for v in values ]
        
        # FIXME: 
        # can we use `np.fromstring('s', dtype=int, sep=',')` here ?
        # it seems that the absent value is translated into `0` instead of NaN                
        Data=pd.DataFrame([map(to_int, v.split(',')) for v in vrows], columns=range(1,366))
        
        # Normalize data (from weather notebook)
        G=Data.ix[:,1:365]
        G[G<-400]=np.NaN
        G[G>500]=np.NaN
        G=G/10
        Data.ix[:,1:365]=G
        
        from numpy import mean, std
        
        # Scale by Mean Std
        def scale_temps(Din):
            matrix=Din.iloc[:,:]
            Dout=Din.loc[:,range(1,366)]
            Mean=mean(matrix, axis=1).values
            Dout['Mean']=Mean
            Std= std(matrix, axis=1).values
            Dout['Std']=Std
            Dout.loc[:,1:365]=matrix.values
            return Dout
        Dout=scale_temps(Data)
        
        # Dout = Scale(Data) + 'mean' + 'std'
        Dout=Dout[['Mean','Std']+range(1,366)]
        
        # Calculate covariance
        M=Dout.loc[:,1:365].transpose()
        M=M.dropna(axis=1)
        (columns,rows)=M.shape
        Mean=mean(M, axis=1).values

        
        C=np.zeros([columns,columns])   # Sum
        N=np.zeros([columns,columns])   # Counter of non-nan entries
        
        for i in range(rows):
            row = M.iloc[:,i]-Mean;
            outer = np.outer(row,row)
            valid = np.isnan(outer)==False
            C[valid] = C[valid]+outer[valid]  # update C with the valid location in outer
            N[valid] = N[valid]+1
        valid_outer = np.multiply(1-np.isnan(N),N>0)
        cov = np.divide(C,N)
        
        # SVD
        U,D,V=np.linalg.svd(cov)
        
        # explained variance
        series = np.cumsum(D[:])/np.sum(D)
        
                
        # Dump some output
        stderr.write('### reducer for group: ' + str(group_id) + ' no rows: ' \
                     + str(len(vrows)) + ', cov: ' + str(cov.shape) + '\n')
        
        yield (group_id, ','.join([str(len(vrows))] + [str(i) for i in series]))
        
        # output format
        #
        # <group_id> "<no_of_measurement>,<e1>,<e2>,...,<e365>"
        #
        # where ei is the explained covariance with i eigenvectors
        
if __name__ == '__main__':
    MRPCA.run()
```
