## Merge groups

```python
def line_arrays(f):
    """
    Parse lines of the form:
    
    <id>  "<int>,<float>,<float>,...,<float>"
    
    """
    import re
    for line in f.readlines():
        matchObj = re.match(r'\s*(\d*)\s*\"(\d*),(\S*)\"\s*', line)
        i = matchObj.group(1)
        n = matchObj.group(2)
        s = matchObj.group(3)
        yield np.array([int(i), int(n)] + [ float(s) for s in s.split(',') ])
```

```python
def get_mergeable_groups(h, filename):
    """
    Parameters
    ----------
    filename: Str
              file with 'explained' variance results from PCA run
    
    Returns
    -------
    list of pairs of groups that need merging
    """
    with open(filename) as f:    
        cols  = ['group_id', 'n']+range(1,366)
        table = [ l for l in line_arrays(f) ]
        df    = pd.DataFrame(table, columns=cols).set_index('group_id')    

        # find how many eignevectors are needed to explained at least 0.99 of the variance
        df['k'] = 365-sum(isnan(df[df<0.99].ix[:,1:365]),axis=1)

        df=df[['n','k']+range(1,366)]

        groups = hi.get_groups()

        n_pairs = [ (i,j) for i in range(len(groups)) 
                          for j in range(i+1,len(groups))
                          if i in groups
                          if j in groups
                          if hi.groups_are_neighbors(i,j) ]

        if len(n_pairs) == 0:
            return []
        
        pdf = pd.DataFrame(n_pairs, columns=['fst','snd'])


        # filling up columns: n1, k1, n2, k2
        dfk = df[['k','n']]
        pdfk = pdf.join(dfk, on='fst').rename(columns = {'k':'k1','n':'n1'})\
                  .join(dfk, on='snd').rename(columns = {'k':'k2','n':'n2'})
        pdfk['k1+k2'] = pdfk['k1'] + pdfk['k2']

        lhs = 'n1*k1+(k1+1)*365 + n2*k2+(k2+1)*365'
        rhs = '(n1+n2)*(k1+k2)+(k1+k2+1)*365'
        
        pdfk[lhs] = pdfk['n1']*pdfk['k1']+(pdfk['k1']+1)*365 \
                   +pdfk['n2']*pdfk['k2']+(pdfk['k2']+1)*365

        pdfk[rhs] = (pdfk['n1']+pdfk['n2'])*(pdfk['k1']+pdfk['k2']) + (pdfk['k1']+pdfk['k2']+1)*365

        
        ## FIXME: Using heuristic here instead of better way...
        
        pdfk['merge'] = pdfk[lhs] / pdfk[rhs] > 0.6

        pdfk = pdfk[pdfk['merge']]
        pdfk = pdfk[['fst','snd']]
        return [ (v['fst'], v['snd']) for (k,v) in pdfk.iterrows() ]       
    
# TODO: fill up k3, we need ....
```

```python
def merge_groups(h, filename):
    """
    Merge groups in hierarchical index `h` based on the results collected
    from PCA on output file `filename`.
    
    Returns the list of newly created groups
    """
    result = []
    for (i,j) in get_mergeable_groups(h, filename):
        new_g = hi.merge_groups(i,j)   
        if new_g >= 0:
            result.append(new_g)
    return result
```

###Loop

```python
# loop condition
cont = True

# loop iterator
itr  = 0

# groups for which PCA has run
computed_groups = set([])

# plot the map
hi.plot_map()   
    

while cont:

    print '================================'
    print 'Iteration', itr
    print '================================'    

    
    # this needs to get updated on every iteration because stations might get assigned to new groups
    grouped_stations = hi.get_grouped_stations()

    
    # dump stations with group information
    pickle_dump(grouped_stations, 'grouped_stations.pklz')
    
    
    # Compute the groups that remain to be computed
    current_groups = grouped_stations['group_id'].unique()
    remaining_groups = list(set(current_groups) - computed_groups)
    
    
    # run the PCA analysis on all groups
    pca_cmd(local_data, key_id, secret_key, 'grouped_stations.pklz', remaining_groups, local_output_file, False)
    #!python pca.py $local_data --key $key_id --secret $secret_key --stations grouped_stations.pklz > $local_output_file
    
    
    #update computed groups
    computed_groups = computed_groups | set(remaining_groups)
    
    # merge where possible
    new_groups = merge_groups(hi,local_output_file)

    
    # plot the map
    hi.plot_map()   
     
    
    # update iterator and loop condition
    if len(new_groups) > 0:
        itr = itr + 1
    else:
        cont = False  

        
print '================================'
print 'Finished'
print '================================'
```
