```python
def line_arrays(f):
    """
    Parse lines of the form:
    
    <id>  "<int>,<float>,<float>,...,<float>"
    
    """
    import re
    for line in f.readlines():
        matchObj = re.match(r'\s*(\d*)\s*\"(\d*),(\S*)\"\s*', line)
        i = matchObj.group(1)
        n = matchObj.group(2)
        s = matchObj.group(3)
        yield np.array([int(i), int(n)] + [ float(s) for s in s.split(',') ])
```

```python
def potential_pairs(h):

    # all existing groups on the map
    groups = h.get_groups()
    group_keys = groups.keys()
    
    def pair_up():
        # the seen nodes
        p = set([])
        for i in group_keys:
            for j in group_keys:
                if i < j and not (i in p) and not (j in p) and h.groups_are_neighbors(i,j):
                    p.add(i)
                    p.add(j)
                    yield (i,j)
                    
    # potential new pairs
    pairs = list(pair_up())
                    
    # first new key that needs to be created
    k = max(groups.keys()) + 1

    # the new tentative group ids
    tmp_group_ids = list(xrange(k, k+len(pairs)+1))
    
    gij = [ (g,i,j) for (g,(i,j)) in zip(tmp_group_ids, pairs) ]
                            
    print 'Potential pairs', gij
    print
    return gij
```

####Possible Optimizaton:
- Memoize results of PCA runs of previous explorations - since not all groups get updated on every run.

```python
# explore new pairings
def explore(h, pairing, input_data, output_dir, emr):
    """
    Parameters
    ----------
    h             : H.I.
    pairing       : list of (int,int,int) that we are going to explore
    input_data    : 
    output_file   : PCA results will end up in `output_file`
    emr           : run on EMR
    
    Returns
    -------
    output_file
    """
    
    # get a temp copy of the group stations  
    tmp_stations = h.get_grouped_stations().copy(deep=True)

    #print 'before adding temp groups'
    #print 'tmp_stations[\'group_id\'].unique()'
    #print tmp_stations['group_id'].unique()
    
    tmp_group_ids = map(lambda (x,y,z): x, pairing)

    for (g,i,j) in pairing:
        tmp_stations.loc[tmp_stations.group_id == i, 'group_id'] = g
        tmp_stations.loc[tmp_stations.group_id == j, 'group_id'] = g 
    
    #print 'after adding temp groups, ie. before pickling'
    #print 'tmp_stations[\'group_id\'].unique()'
    #print tmp_stations['group_id'].unique()

    
    # dump stations with group information
    pickle_dump(tmp_stations, 'grouped_stations-tmp.pklz')

    # output file
    output_file = output_dir + 'pca_local_tmp.txt'
    
    pca_cmd(input_data, 'grouped_stations-tmp.pklz', tmp_group_ids, output_file, emr)
    
    return output_file
```

```python
def get_mergeable_groups(pairing, main_results, exploration_results):
    """
    Parameters
    ----------
    Each file contains lines of the form:
        
    <id>  "<int>,<float>,<float>,...,<float>"
      
    Returns
    -------
    list of (int,int) : a list of group ids that need merging
    """
    
    # vector length
    vlen = 365
    
    # results from main run
    
    table = [ l for l in line_arrays(open(main_results)) ]
    cols  = ['group_id', 'n']+range(1,vlen+1)
    
    mdf    = pd.DataFrame(table, columns=cols).set_index('group_id')    

    # find how many eignevectors are needed to explain at least 0.99 of the variance
    mdf['k'] = vlen-sum(isnan(mdf[mdf<0.99].ix[:,1:vlen]),axis=1)

    
    # results from exploration run
    
    table = [ l for l in line_arrays(open(exploration_results)) ]
    cols  = ['group_id', 'n3']+range(1,vlen+1)
    
    edf = pd.DataFrame(table, columns=cols).set_index('group_id')    

    edf['i']  = -1
    edf['j']  = -1    
    edf['n1'] = -1 
    edf['n2'] = -1 

    # find how many eignevectors are needed to explain at least 0.99 of the variance
    edf['k3'] = vlen-sum(isnan(edf[edf<0.99].ix[:,1:vlen]),axis=1)
        
    pairs_mapping = {}
    for (g,i,j) in pairing:
        edf.ix[g,'i'] = i
        edf.ix[g,'j'] = j        
        edf.ix[g,'n1'] = mdf.ix[i,'n']
        edf.ix[g,'n2'] = mdf.ix[j,'n']
        edf.ix[g,'k1'] = mdf.ix[i,'k']
        edf.ix[g,'k2'] = mdf.ix[j,'k']
        
    
    lhs = 'n1*k1+(k1+1)*365 + n2*k2+(k2+1)*365'
    rhs = 'n3*k3+(k3+1)*365'

    edf[lhs] = edf['n1']*edf['k1']+(edf['k1']+1)*vlen \
             + edf['n2']*edf['k2']+(edf['k2']+1)*vlen

    edf[rhs] = edf['n3']*edf['k3']+(edf['k3']+1)*vlen
            
    edf['merge'] = edf[lhs] >= edf[rhs] 
    
    # print edf[['i','j','n3','k3','n1','k1','n2', 'k2', lhs, rhs, 'merge']]

    return [ (v['i'], v['j']) for (k,v) in edf.iterrows() if v['merge'] ]
```

```python
def merge_groups(h, input_data, main_results, output_dir, emr):
    """
    Merge groups in hierarchical index `h` based on the results collected
    from PCA on output file `filename`.
    
    Returns the list of newly created groups
    """
    
    # Get pairs of groups that can potentially be merged (are neighboring)
    pairing = potential_pairs(h)
    
    # Use this file to dump the exploration output
    exploration_results = output_dir + 'pca_local_tmp.txt'    
    
    # Do the exploration (run the PCA)
    explore(h, pairing, input_data, output_dir, emr)
    
    # Perform the MDL criterion to chose merge pairs
    merge_groups = get_mergeable_groups(pairing, main_results, exploration_results)
    
    result = []
    for (i,j) in merge_groups:
        new_g = h.merge_groups(i,j)   
        if new_g >= 0:
            print 'Merging groups:',i,j,'-->',new_g
            result.append(new_g)
            
    return result
```

```python
def run (num_levels, output_dir, input_data, emr):
    """
    
    num_levels  : the number of levels of the index tree
    output_dir  : e.g. '/home/ubuntu/panagosg7/notebooks/weather.mapreduce/output/'
    input_data  : e.g. '/home/ubuntu/data/weather/SAMPLE_TMAX.csv'
    
    """
    
    # the dataframe containing stations with 'weight' information
    weighted_stations = pickle.load(gzip.open( 'weighted_stations.pklz', "rb" ))

    # the hierarchical index
    hi = HierarchicalIndexing(weighted_stations, num_levels)
    
    # local output file
    if emr: 
        output_file = output_dir + 'pca_emr.txt'
    else:    
        output_file = output_dir + 'pca_local.txt'
    
    !rm -f $output_file
    

    # input data
    print 'Input data:', input_data
    
    # loop condition
    cont = True

    # loop iterator
    itr  = 0

    # groups for which PCA has run
    computed_groups = set([])


    while cont:

        print '================================'
        print 'Iteration', itr
        print '================================'    

        # plot the map
        hi.plot_map('Running PCA on these groups')


        # this needs to get updated on every iteration because stations might get assigned to new groups
        grouped_stations = hi.get_grouped_stations()
        
        print 'GROUPED STATIONS AT START OF ITR: ', itr
        print grouped_stations['group_id'].unique()


        # dump stations with group information
        pickle_dump(grouped_stations, 'grouped_stations.pklz')


        # Compute the groups that remain to be computed
        current_groups = grouped_stations['group_id'].unique()
        remaining_groups = list(set(current_groups) - computed_groups)
         

        # run the PCA analysis on all groups
        # XXX: we shouldn't overwrite the old results we want everything there!
        pca_cmd(input_data, 'grouped_stations.pklz', remaining_groups, output_file, emr, overwrite=False)
        #!python pca.py $local_data --key $key_id --secret $secret_key --stations grouped_stations.pklz > $local_output_file
        

        # update computed groups
        computed_groups = computed_groups | set(remaining_groups)

        
        # merge where possible
        new_groups = merge_groups(hi, input_data, output_file, output_dir, emr)


        # update iterator and loop condition
        if len(new_groups) > 0:
            itr = itr + 1
        else:
            cont = False  


    print '================================'
    print 'Finished'
    print '================================'
```
