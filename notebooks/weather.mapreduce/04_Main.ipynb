```python
import re

def line_arrays(f):
    for line in f.readlines():
        matchObj = re.match(r'\s*(\d*)\s*\"(\d*),(\S*)\"\s*', line)
        i = matchObj.group(1)
        n = matchObj.group(2)
        s = matchObj.group(3)
        yield np.array([int(i), int(n)] + [ float(s) for s in s.split(',') ])
    
def get_mergeable_groups(filename):
    """
    Parameters
    ----------
    filename: Str
              file with 'explained' variance results from PCA run
    
    Returns
    -------
    list of pairs of groups that need merging
    """
    with open(filename) as f:    
        cols  = ['group_id', 'n']+range(1,366)
        table = [ l for l in line_arrays(f) ]
        df    = pd.DataFrame(table, columns=cols).set_index('group_id')    

        # find how many eignevectors are needed to explained at least 0.99 of the variance
        df['k'] = 365-sum(isnan(df[df<0.99].ix[:,1:365]),axis=1)

        df=df[['n','k']+range(1,366)]

        groups = hi.get_groups()

        n_pairs = [ (i,j) for i in range(len(groups)) 
                          for j in range(i+1,len(groups))
                          if i in groups
                          if j in groups
                          if hi.groups_are_neighbors(i,j) ]

        pdf = pd.DataFrame(n_pairs, columns=['fst','snd'])


        # filling up columns: n1, k1, n2, k2
        dfk = df[['k','n']]
        pdfk = pdf.join(dfk, on='fst').rename(columns = {'k':'k1','n':'n1'})\
                  .join(dfk, on='snd').rename(columns = {'k':'k2','n':'n2'})
        pdfk['k1+k2'] = pdfk['k1'] + pdfk['k2']

        lhs = 'n1*k1+(k1+1)*365 + n2*k2+(k2+1)*365'
        rhs = '(n1+n2)*(k1+k2)+(k1+k2+1)*365'
        
        pdfk[lhs] = pdfk['n1']*pdfk['k1']+(pdfk['k1']+1)*365 \
                   +pdfk['n2']*pdfk['k2']+(pdfk['k2']+1)*365

        pdfk[rhs] = (pdfk['n1']+pdfk['n2'])*(pdfk['k1']+pdfk['k2']) + (pdfk['k1']+pdfk['k2']+1)*365

        pdfk['merge'] = pdfk[lhs] / pdfk[rhs] > 0.6

        pdfk = pdfk[pdfk['merge']]
        pdfk = pdfk[['fst','snd']]
        return [ (v['fst'], v['snd']) for (k,v) in pdfk.iterrows() ]       
    
# TODO: fill up k3, we need ....
```

## Merge groups

```python
def merge_groups(h, filename):
    """
    Merge groups in hierarchical index `h` based on the results collected
    from PCA on output file `filename`.
    """
    return [ new_g for (i,j) in get_mergeable_groups(filename) for new_g in [hi.merge_groups(i,j)] if new_g > 0 ]
```

## Draw a random rectangle `i` and all its neighbors

```python
# http://matplotlib.org/basemap/users/merc.html
from mpl_toolkits.basemap import Basemap
import numpy as np
import random
import matplotlib.pyplot as plt

def plot_map(hi):
    lonmin=-180;lonmax=180;latsmin=-80;latsmax=80;
    plt.figure(figsize=(15,10),dpi=300)
    m = Basemap(projection='merc', 
                llcrnrlat=latsmin,urcrnrlat=latsmax,\
                llcrnrlon=lonmin,urcrnrlon=lonmax, \
                lat_ts=20,
                resolution='c')  # 'l','i'
    m.drawcoastlines()
    m.fillcontinents(color='#E2E2E2',lake_color='#FFFFFF')
    m.drawmapboundary(fill_color='#FFFFFF')


    rectangles = hi.get_rectangles()
    groups = hi.get_groups()

    for group_id in groups:
        group = groups[group_id]
        rect_ids = group['rects']
        # use common color for same group
        r = lambda: random.randint(0,255)
        color = '#%02X%02X%02X' % (r(),r(),r())
        for rect_id in rect_ids:
            rect = rectangles.ix[rect_id]
            draw_screen_poly(plt, group_id, rect.west, rect.east, rect.south, rect.north, m, color)

    # plt.legend()
    plt.title('World groups')
    plt.show()
```
